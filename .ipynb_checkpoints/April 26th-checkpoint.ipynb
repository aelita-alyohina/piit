```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
df=data('mtcars')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#1. Histogram
plt.figure(figsize=(15,5))
plt.hist(df['mpg'])
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
#MPG distribution: Mostly clustered between ~18-24 typical mid-efficiency cars.
```

```python
#2. Scatter
plt.figure()
plt.scatter(df['hp'], df['mpg'])
plt.title('HP vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()
```

```python
#3. Boxplot
plt.figure()
df.boxplot(column='mpg', by='cyl')
plt.title('MPG by Cylinders')
plt.suptitle('')
plt.show()
```

```python
#4. Line Plot
df_sorted=df.sort_values(by='wt')
plt.figure()
plt.plot(df_sorted['wt'], df_sorted['mpg'], color='r')
plt.title("Weight vs MPG")
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```
