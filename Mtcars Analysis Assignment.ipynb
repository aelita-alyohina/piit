## Part 4

```python
#12 Data Visualization using matplot
```

```python
import matplotlib.pyplot as plt

df.groupby('cyl')['mpg'].mean().plot(kind='bar')

plt.title('Average mpg by cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Mpg')
plt.show()
```

```python
#Cylinders and mpg are inversely proportional. Cars with less cylinders have higher mpg.
```

```python
plt.scatter(df['wt'],df['mpg'])
plt.title('Weight vs Mpg')
plt.xlabel('Weight')
plt.ylabel('Mpg')
plt.show()
        
```

```python
#Heavier cars have less mpg.
```

```python
plt.hist(df['hp'])
plt.title('Horsepower Distribution')
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.show()
```

```python
#Most cars are not too powerful. Majority are everyday cars, and only small amount of cars are very powerful ones.
```

```python
df.boxplot(column='mpg', by='am')
plt.title('Mpg by Transmission')
plt.suptitle('')
plt.xlabel('Transmission(0=automatic, 1=manual)')
plt.ylabel('Mpg')
plt.show()
```

```python
#Manual transmission cars have higher median mpg than automatic cars which makes them more fuel efficient.
```
